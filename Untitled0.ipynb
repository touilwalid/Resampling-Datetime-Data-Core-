```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
df = df[(df['date'] >= '2000-01-01') & (df['date'] <= '2010-12-31')]
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(df['date'], df['precipitation'], color='blue')
ax.set_title('Precipitations for 2000-2010')
ax.set_xlabel('Date')
ax.set_ylabel('Pouces')
ax.xaxis.set_major_locator(mdates.YearLocator(1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=90)
ax.xaxis.set_minor_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_minor_formatter(mdates.DateFormatter(''))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
annual_mean_temp = df.resample('Y', on='date')['mean_temp'].mean()
coolest_year = annual_mean_temp.idxmin()
coolest_temp = annual_mean_temp.min()
plt.figure(figsize=(10, 6))
plt.plot(annual_mean_temp.index, annual_mean_temp.values, color='blue')
plt.axvline(x=coolest_year, color='red', linestyle='--', label=f'Coolest Year ({coolest_year.year}) - {coolest_temp}°C')
plt.legend()
plt.title('Average Yearly Temperature')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.grid(True)
plt.tight_layout()
plt.show()
print("Year with the coolest average temperature:", coolest_year.year)
print("Lowest mean temperature value:", coolest_temp)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
annual_mean_temp = df.resample('Y', on='date')['mean_temp'].mean()
coolest_year = annual_mean_temp.idxmin()
coolest_temp = annual_mean_temp.min()
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(annual_mean_temp.index, annual_mean_temp.values, color='blue')
ax.axvline(x=coolest_year, color='red', linestyle='--', label=f'Coolest Year ({coolest_year.year}) - {coolest_temp}°C')
ax.legend()
ax.set_title('Average Temperature')
ax.set_ylabel('Degrees')
ax.xaxis.set_major_locator(mdates.YearLocator(5))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=90)
ax.xaxis.set_minor_locator(mdates.YearLocator())
ax.xaxis.set_minor_formatter(mdates.DateFormatter(''))
ax.grid(True)
fig.autofmt_xdate()
plt.tight_layout()
plt.show()
print("Year with the coolest average temperature:", coolest_year.year)
print("Lowest mean temperature value:", coolest_temp)
```
